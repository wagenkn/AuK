```python
def find_index(iterable, value):
    for i in range(len(iterable)):
        if iterable[i] == value:
            return i


def car_parking(src, tgt):
    operations = []
    for i in range(len(tgt)):
        if src[i] != tgt[i]:
            if src[i] != 0:
                other_index = find_index(src, 0)
                operations.append((src[i], 0))
                src[i], src[other_index] = src[other_index], src[i]
            other_index = find_index(src, tgt[i])
            operations.append((src[other_index], 0))
            src[i], src[other_index] = src[other_index], src[i]
    return operations
```
