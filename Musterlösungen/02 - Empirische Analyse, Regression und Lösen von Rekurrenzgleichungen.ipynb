```python
import random
import statistics
import time
import matplotlib.pyplot as plt
import numpy as np


def max_heapify(heap, i):
    if i < len(heap)/2:
        if heap[2*i-1] > heap[i-1] and heap[2*i-1] > heap[2*i]:
            heap[i-1], heap[2*i-1] = heap[2*i-1], heap[i-1]
            return max_heapify(heap, 2*i)
        elif heap[2*i] > heap[i-1]:
            heap[i-1], heap[2*i] = heap[2*i], heap[i-1]
            return max_heapify(heap, 2*i+1)
    return heap


def build_max_heap(arr):
    for i in reversed(range(1, int(len(arr)/2) + 1)):
        arr = max_heapify(arr, i)
    return arr


def get_random_lst(size, min_value=0, max_value=100000):
    lst = []
    for i in range(size):
        lst.append(random.randint(min_value, max_value))
    return lst


def measure_time(algorithm):
    def helper(*args):
        start = time.clock()
        algorithm(*args)
        return time.clock() - start

    return helper


def get_sum_x_y(x, y):
    a = 0
    for i in range(len(x)):
        a += x[i] * y[i]
    return a


def linear_regression_function(x, y):
    assert len(x) == len(y)
    mean_x = statistics.mean(x)
    mean_y = statistics.mean(y)
    a = (get_sum_x_y(x, y) - len(x) * mean_x * mean_y) / (sum(map(lambda x: x * x, x)) - len(x) * mean_x ** 2)
    b = mean_y - a * mean_x
    return lambda x: a * x + b


def covariance(x, y):
    cov = 0
    mean_x = statistics.mean(x)
    mean_y = statistics.mean(y)
    for i in range(len(x)):
        cov += (x[i] - mean_x) * (y[i] - mean_y)
    return cov / len(x)


def correlation_coefficient(x, y):
    assert len(x) == len(y)
    return covariance(x, y) / (statistics.pstdev(x) * statistics.pstdev(y))


x = list(map(lambda n: 2 * n * 10 ** 5, list(range(1, 6))))
y = list(map(lambda x: statistics.median(list(map(lambda _: measure_time(build_max_heap)(get_random_lst(x)), range(1)))), x))

linear_regression_function(x, y)

plt.scatter(x, y)

helper_x_points = np.linspace(0, 1.1 * max(x), 1000)
plt.plot(helper_x_points, linear_regression_function(x, y)(helper_x_points))

plt.xlabel('problem size n')
plt.ylabel('elapsed time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()

print('r={}'.format(correlation_coefficient(x, y)))
```
