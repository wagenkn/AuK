# Komplexitätstheorie und NP-vollständige Probleme

#### 1. 

> a) Mit $X \leqslant_p Y$ zeigt man, dass $Y$ mindestens so schwer wie $X$ ist. Um die NP-Schwere von $X$ zu zeigen, muss man aber zeigen, dass $X$ mindestens so schwer ist wie ein NP-vollständiges Problem $Y$.

> D.h., um zu zeigen, dass das Problem $X$ NP-vollständig ist, reicht es aus zu zeigen, dass $X$ in NP liegt und $Y \leqslant_p X$ für ein NP-vollständiges Problem $Y$.

> b) Da $X$ ein NP-vollständiges Problem ist, ist es mindestens so schwer wie das SAT-Problem. Gibt es für das SAT-Problem keinen effizienten Algorithmus, so gibt es auch keinen für das Problem $X$.

> Es gibt kein NP-vollständiges Problem $X$, das mit einem effizienten deterministischen Algorithmus in Polynomialzeit gelöst werden kann, wenn es keinen deterministischen Polynomialzeit-Algorithmus für das SAT-Problem gibt.

> c) Mit $X \leqslant_p SAT$ zeigt man, dass $SAT$ mindestens so schwer wie $X$ ist. Da $X$ NP-schwer ist, ist es mindestens so schwer wie ein NP-vollständiges Problem. Das SAT-Problem ist NP-vollständig. Deshalb lautet es korrekt:

> Das SAT-Problem kann auf jedes NP-schwere Problem in Polynomialzeit reduziert werden.

#### 2.

> a) Angenommen der Algorithmus ruft eine Folge von Subroutinen $S_1, S_2, \dotsc, S_m$ auf. Da jede Subroutine eine andere Subroutine in ihr selber aufrufen kann, wird der Zeitwaufwand des Algorithmus am größten, wenn für alle Subroutinen die nächste in ihr selber aufgerufen wird. $S_{i+1}$ wird also in $S_i$ aufgerufen. Da die Subroutinen alle in Polynomialzeit laufen, lassen sie sich durch ein Polynom $p(n)$ nach oben beschränken.

> Der Zeitaufwand $T$ des Algorithmus ist dann:

> $$
\begin{align*}
T(m) &= p(p( \dotsc p(n) \dotsc ))
\end{align*}
$$

> Die Anzahl der Aufrufe enstpricht $m$. Für $T(m)$ ergibt sich folgende Darstellung:

> $$
\begin{align*}
T(m) &= p(n) \cdot T(m-1) \\
T(m) &= p(n) \cdot p(n) \cdot T(m-2) \\
T(m) &= p^2(n) \cdot T(m-2) \\
T(m) &= p^i(n) \cdot T(m-i) \\
\end{align*}
$$

> Da $T(0)$ eine Konstante $c$ ist, muss $T(m-i)$ gleich $T(0)$ gesetzt werden.

> $$
\begin{align*}
T(m-i) = T(0) & \implies m-i = 0 \implies m = i \\
T(m) &= p^m(n) \cdot T(0) \\
T(m) &= c \cdot p^m(n)
\end{align*}
$$

> Da nur konstant viele Subroutinen aufgerufen werden, ist $m$ eine Konstante. Dadurch ist $p^m(n)$ ist immer noch ein Polynom. Der Algorithmus hat also polynomiellen Aufwand.
<div style="text-align: right; font-size: 24px;">&#9633;</div>

> b) Betrachten wir den Ausdruck $T(m) = c \cdot p^m(n)$ aus Teilaufgabe a) und setzen $m$ zu etwas Polynomialem, z.B. $m = n$, so entsteht der Ausdruck $T(m) = c \cdot p^n(n)$. Jetzt befindet sich $n$ im Exponenten und es handelt sich somit um exponentiellen Aufwand.

#### 3.

> a) Gibt es eine Teilmenge eines Graphen $G = (V, E)$ mit $k$ Knoten, sodass jede Kante an einen Knoten dieser Teilmenge angrenzt?

> b) __Vertex Cover ist in NP__:

>   Man kann raten, ob ein Knoten zum *Vertex Cover* gehört. Es gibt polynomiell viele Knoten, damit liegt das Problem in NP.

>   __Vertex Cover ist NP-schwer__:
    
>   Ist $S$ ein Independent Set von $G = (V, E)$, so ist $(V - S)$ Vertex Cover.

>   Weniger formal formuliert, muss man jeden Knoten invertieren, d.h. jeder Knoten, der nicht zur Teilmenge von *Independent Set* gehört, gehört zur Teilmenge von *Vertex Cover* und jeder Knoten, der zur Teilmenge von *Independent Set* gehört, dehört nicht zur Teilmenge von *Vertex Cover*. Die Größe des *Vertex Cover* ist $\left| V \right| - \left| S \right| = \left| V \right| - k$, wobei $k$ die Lösung von *Independent Set* ist. $\left| V \right| - k$ ist dann entsprechend die Lösung von *Vertex Cover*.

>   Dies funktioniert, da bei *Independent Set* jede Kante an maximal einen Knoten, der zur Teilmenge gehört, angrenzt. Jede Kante grenzt also an entweder einen oder keine Knoten, der zur Teilmenge gehört. Invertiert man nun jeden Knoten, so grenzt jede Kante an entweder einen oder zwei, also mindestens einen Knoten, der zur Teilmenge gehört. Durch diese Tatsache, handelt es sich um einen *Vertex Cover*.

>   Damit ist eine Reduktion gegeben und die NP-Vollständigkeit von *Vertex Cover* bewiesen.
